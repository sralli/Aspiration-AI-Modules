# Query 2.5

Plot the daily return (percentage) distribution as a histogram.
Histogram analysis is one of the most fundamental methods of exploratory data analysis. In this case, it'd return a frequency plot of various values of percentage changes .

```python
df.head()
```

```python
def calculate_daily_returns(df):
  daily_returns = (df['Close Price'] / df['Close Price'].shift(1)) -1
  daily_returns.ix[0] = 0
  return daily_returns

def plot_histogram(df, bins):
  mean = df.mean(skipna=True)
  std = df.std(skipna=True)
  df.hist(bins=bins)
  plt.axvline(mean.item(), color='w', linestyle='dashed', linewidth=2)
  plt.axvline(std.item(), color='r', linestyle='dashed')
  plt.axvline(-std.item(), color='r', linestyle='dashed')
  plt.show()
```

```python
calculate_daily_returns(df)
```

# Query 2.9

Trade Calls - Using Simple Moving Averages.

Plot the 21 day and 34 day Moving average with the average price and decide a Call ! 
Call should be buy whenever the smaller moving average (21) crosses over longer moving average (34) AND the call should be sell whenever smaller moving average crosses under longer moving average. 
One of the most widely used technical indicators.

## Using INFY Stock

```python
#Attempting SImple Moving Averages: 

rolling_mean = df['Close Price'].rolling(21).mean()
rolling_mean2 = df['Close Price'].rolling(34).mean()
```

```python
plt.figure(figsize=(20,20))
plt.plot(df.Date, df['Close Price'], label='INFY')
plt.plot(df.Date, rolling_mean, label='Infosys 21 Day SMA', color='orange')
plt.plot(df.Date, rolling_mean2, label='Infosys 34 Day SMA', color='magenta')
plt.legend(loc='upper left')
plt.show()
```

```python
for q in range(len(rolling_mean2)):
    if rolling_mean[q]>rolling_mean2[q] and rolling_mean[q-1]<rolling_mean2[q-1]:
        print("Call: Buy at", df['Close Price'][q])
    elif rolling_mean[q]<rolling_mean2[q] and rolling_mean[q-1]>rolling_mean2[q-1]:
        print("Call: Sell at", df['Close Price'][q])
```

# Query 2.10

Trade Calls - Using Bollinger Bands 

Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average 

The bollinger bands comprise the following data points- 
- The 14 day rolling mean of the closing price (we call it the average) 
- Upper band which is the rolling mean + 2 standard deviations away from the average. 
- Lower band which is the rolling mean - 2 standard deviations away from the average. 
- Average Daily stock price.

Bollinger bands are extremely reliable , with a 95% accuracy at 2 standard deviations , and especially useful in sideways moving market.

Observe the bands yourself , and analyse the accuracy of all the trade signals provided by the bollinger bands. 

```python
def Bollinger(df, n):
    MA = df['Close Price'].rolling(window=n).mean()
    SD = df['Close Price'].rolling(window=n).std()
    df['UpperBB'] = MA + (2*SD)
    df['LowerBB'] = MA - (2*SD)
    return df

n =14
INFY_BBANDS = Bollinger(df, n)
```

```python
plt.figure(figsize=(20,20))
plt.plot(df.Date, df['Close Price'].rolling(14).mean(), label = 'Rolling 14 Day Price', color = 'Red')
plt.plot(df.Date, df['Close Price'], label='INFY Average Daily Stock Price', color = 'Black', linewidth=2.0)
df['MiddleBB'] = df['Close Price'].rolling(14).mean()
plt.plot(df.Date, df.UpperBB, label='Upper Bollinger Band', color='orange')
plt.plot(df.Date, df.LowerBB, label='Lower Bollinger Band', color='magenta')

plt.legend(loc='upper left')
plt.show()

```

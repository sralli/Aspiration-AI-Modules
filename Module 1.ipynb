```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
def vwap(df):
    q = df['Total Traded Quantity'].values
    p = df['Close Price'].values
    return df.assign(vwap=(p * q).cumsum() / q.cumsum())
```

```python

df = df.groupby([df.Year, df.Month],group_keys = False).apply(vwap)
```

```python
#Task 5
def average(n):
    return df['Close Price'][-n:].mean()

def profit_loss(n):
    return ((df['Close Price'][-n:-n+1]/df['Close Price'][-1]))*100
```

```python
# 1 WEEK
print(average(7))
print(profit_loss(7))
```

```python
#Task 1.6

df['Day_Perc_Change'] = df['Close Price'].pct_change()
```

```python
df['Day_Perc_Change'][0] = 0
```

```python
df.head()
```

```python
#Task 1.7
Temp = []
for i in df['Day_Perc_Change']: 
    if i>-0.5 and i<=0.5:
        Temp.append("Slight or No change")
    elif i>0.5 and i<=1:
        Temp.append("Slight positive")
    elif i>-1 and i<=-0.5:
        Temp.append("Slight negative")
    elif i>1 and i<=3:
        Temp.append("Positive")
    elif i>-3 and i<=-1:
        Temp.append("Negative")
    elif i>=-7 and i<=-3:
        Temp.append("Among top losers")
    elif i>3 and i<=7:
        Temp.append("Among top gainers")
    elif i>7:
        Temp.append("Bull run")
    elif i<-7:
        Temp.append("Bear drop")
```

```python
df['Trend'] = Temp
```

```python
df.head()
```

```python
group_by_Trend = df.groupby(['Trend', 'Total Traded Quantity'])
```

```python
#Task 1.8 average
print(group_by_Trend.mean())
```

```python
#Task 1.8 median

print(group_by_Trend.median())
```
